## Import Libraries
- Seaborn automatically install these libraries
- numpy
- scipy
- pandas
- matplotlib

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
phool = sns.load_dataset("iris")
phool

#  Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
# Add coordinates limits
plt.xlim(4.1)
plt.ylim(2)
# Add title
plt.title("Phoolon Ka Plot")
plt.show()
```

## Set Style
- darkgrid  
- whitegrid  
- dark  
- white  
- ticks

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import plotly

# Load dataset
phool = sns.load_dataset("iris")

# Change figure size
plt.figure(figsize=(12,6))

#  Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
# Add coordinates limits
plt.xlim(4.25)
plt.ylim(2.24)

#  Set style
sns.set_style("whitegrid")
# clear old setstyle first
sns.set_style(style=None, rc=None)
# Add title
plt.title("Phoolon Ka Plot")
plt.show()
```
