```python
data = pd.read_csv('pk-frtlizer.csv')
plt.figure(figsize=(40,15))
sns.barplot(x="Year", y="Value", hue="Item", data=data,
           palette="Set1")

plt.show()
```

```python
data = pd.read_csv('pk-frtlizer.csv')
plt.figure(figsize=(40,15))
sns.lineplot(x="Year", y="Value", hue="Item", data=data,
           palette="Set1")

plt.show()
```

```python
data = pd.read_csv('pk-frtlizer.csv')
plt.figure(figsize=(40,15))
sns.boxplot(x="Year", y="Value", hue="Item", data=data,
           palette="Set1")

plt.show()
```

```python
data = pd.read_csv('pk-frtlizer.csv')
sns.set(style='whitegrid')
plt.figure(figsize=(40,15))
sns.lmplot(x="Year", y="Value", hue="Item", data=data,
           palette="Set1")

plt.show()
```

```python
data = pd.read_csv('pk-frtlizer.csv')
plt.figure(figsize=(40,15))
sns.kdeplot(x="Year", y="Value", hue="Item", data=data,
           palette="Set1")

plt.show()
```
